```python
import torch
from transformers import pipeline
import os
```

```python
dolly_pipeline = pipeline(model="databricks/dolly-v2-3b",
                            torch_dtype=torch.bfloat16,
                            trust_remote_code=True,
                            device_map="auto")
```

```python
def get_completion_dolly(input):
  system = f"""
  You are an excellent Physicist.
  You are good at making a lay man understand complex problems.
  Help as much as you can.
  """
  prompt = f"#### System: {system}\n#### User: \n{input}\n\n#### Response from Dolly-v2-3b:"
  print(prompt)
  dolly_response = dolly_pipeline(prompt,
                                  max_new_tokens=500
                                  )
  return dolly_response[0]["generated_text"]
```

```python
prompt = "What is a boson."
print(get_completion_dolly(prompt))
```

```python
import gradio as gr
interface = gr.Interface(fn=get_completion_dolly, inputs=[gr.Textbox(label="Your question",lines=6)],
                     outputs=[gr.Textbox(label="Answer",lines=3)],
                     title="Physics Assistant",
                     examples=["Explain the collision of sub atomic particles.",
                              "How does space and time work"]
                     )
interface.launch(share=True)
```
